```python
from functions import sigmoid
import numpy as np
import matplotlib.pyplot as plt
```

```python
comf_dist = 3.0
dists = np.linspace(0.01, comf_dist**2, 100)
def f(x, fac = 1.2):
    return (sigmoid((comf_dist / 2 - x) * fac) * x + 0.5*x) / 1.5
fs = lambda fac: [f(x, fac) for x in dists]

plt.plot(dists, fs(0.8))
plt.plot(dists, fs(1.0))
plt.plot(dists, fs(1.25))
plt.plot(dists, fs(1.5))
plt.plot(dists, fs(1.75))
plt.plot(dists, fs(2.0))
```
